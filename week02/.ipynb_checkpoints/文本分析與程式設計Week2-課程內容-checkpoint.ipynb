## 如何利用詞性來取出特徵詞
- 我們使用 Articut 內建的 .getNounStemLIST() 函式來取得文本中各句的名詞。

## 關於 .getNounStemLIST() 小補充
- 請看以下程式範例
- .getNounStemLIST() 中要放入前面由articut 斷好詞的結果 (在這裡是result)，另外一個參數可以調整的是indexWithPOS。如果是True 會包含在result_pos 裡面所有字符都會計算到，這個參數的預設也是True。
- 在indexWithPOS = True 的情況之下，在使用.getNounStemLIST()後，這個函數會回傳類似(47, 49, '總統')的內容
- 其中，47 是代表在句子中的位置，只是這個數字是連前面POS 的結果的字符都有納入計算，所以例如我們看「\<ENTITY_person\>蔡英文\<\/ENTITY_person\>\<ENTITY_noun\>總統」這段，"<"是第0個，"E"是第一個，以此類推，總統是第47個，而要表達總統二字，用python list的表示方法是47:49。(python 的47:49就是第47和48個的意思) 

```python
inputSTR = "蔡英文總統明日到台北市政府找柯文哲開會討論他的想法，請你安排一下！"

result = articut.parse(inputSTR)
nounStemLIST = articut.getNounStemLIST(result, indexWithPOS=True)
print(result['result_pos']) #為了輔助解釋，所以特別看result 中 result_pos 的部分
print(nounStemLIST)
```

- 而在indexWithPOS = True 的情況之下，在使用.getNounStemLIST()後，這個函數會回傳類似(3, 5, '總統')的內容
- 因為在「蔡英文總統明日到台北市政府找柯文哲開會討論他的想法，請你安排一下！」中，「蔡」是第0個字，「英」是第1個字，「文」是第2個字，而「總統」分別是第3和第4個字，在python list的表示方法是第3:5個

```python
result = articut.parse(inputSTR)
nounStemLIST = articut.getNounStemLIST(result, indexWithPOS=False)
print(result['result_pos']) #為了輔助解釋，所以特別看result 中 result_pos 的部分
print(nounStemLIST)
```

- 因為我們這邊不需要知道這些字在句子裡面的位置，所以我們就寫一個函式，來只取出字詞就好了

```python
def wordExtractor(inputLIST):
    '''
    配合 Articut() 的 .getNounStemLIST() 和 .getVerbStemLIST() 拋棄位置資訊，只抽出詞彙。
    '''
    resultLIST = []
    for i in inputLIST: #在所有inputLIST裡面的小LIST 中
        if i == []:     #如果list是空的
            pass        #就忽略
        else:
            for e in i:                  #在小list裡面的元素
                resultLIST.append(e[-1]) #把內容存在resultLIST中
    return sorted(list(set(resultLIST))) #把所有的內容排序(sort)、放在list 裡面 (list)、用set排除重複的內容 (set)
```

```python
## 名詞
baseballNounLIST = articut.getNounStemLIST(baseballResultDICT) 
print(wordExtractor(baseballNounLIST))

basketballNounLIST = articut.getNounStemLIST(basketballResultDICT)
print(wordExtractor(basketballNounLIST))
```

```python
## 動詞
baseballVerbLIST = articut.getVerbStemLIST(baseballResultDICT) 
print(wordExtractor(baseballVerbLIST))

basketballVerbLIST = articut.getVerbStemLIST(basketballResultDICT)
print(wordExtractor(basketballVerbLIST))
```
