### 進階任務: 
在這個任務中，我們希望自己透過對詞性標記的觀察，留下標記為UserDefined,ACTION,ENTITY,VerbP，並且只留下句子(個人定義句子為有Entity/UserDefined跟Action)，希望透過這樣子的方式我們可以過濾目前不需要的元素，但是又保持句子的概念，如此一來，當我們發現"股價"跟"走揚"出現在同一個句子時，我們就可以比較放心認定他是一個有關股市上漲的句子了，話不宜遲，我們馬上開始吧!

```python
#先觀察POS的形式
stockResultDICT["result_pos"][:20]
```

```python
#只留下UserDefined,ACTION,ENTITY,VerbP，並且只留下句子，個人定義句子為有Entity/UserDefined跟Action
# step1: 找句子
import re
sentenceLIST = []
for i in stockResultDICT["result_pos"]:
    if ("ENTITY" in i or "UserDefined" in i) and ("ACTION" in i):
        sentenceLIST.append(i)
sentenceLIST[:2]
```

```python
import re
# step2: 將UserDefined,ACTION,ENTITY留下，將每句存成一個LIST，並移除POS
def selectNVLIST(inputSTR):
    wordLIST = re.findall("<ACTION[^>]*>[^<]*</ACTION[^>]*>|<ENTITY[^>]*>[^<]*</ENTITY[^>]*>|<UserDefined>[^<]*</UserDefined>|<VerbP>[^<]*</VerbP>",inputSTR)
    newWordLIST = []
    for i in wordLIST:
        word = re.sub(r"<[^>]*>","",i)
        newWordLIST.append(word)
    return newWordLIST
```

```python
stockNVLIST = [selectNVLIST(i) for i in sentenceLIST]
stockNVLIST 
# 1. 近期股價持續走揚
# 2. 漲幅達9%，近期有望挑戰歷史最高價85.6元。
# 3. 年增26.5%
# 4. 都讓元太今年客戶端的需求大幅成長
# 5. 將使元太下半年業績逐步升溫


```

在這個任務中，我們開始要將我們在思考時間所累積的想法實踐出來，我們首先要先藉由人工觀察，建出一個有關於描述股市漲的LIST，之後將它和新聞去進行比對以得到結果。

```python
#方便起見，我們都先將這些文本存成一個LIST
stockContentLIST = list(stockDF["Content"])
stockContentLIST[1]
```

```python
#觀察資料建立描述漲的詞彙
goUpTargetLIST = ["市場","股價","買盤","指數","漲幅","需求","營收","營運","同期","出貨","獲利"]
goUpActionLIST = ["看好","漲","湧入","登上","增","成長", "升","提升"]
```

```python
#以最後一個方法為例:留下UserDefined,ACTION,ENTITY,VerbP，並且只留下句子(個人定義句子為有Entity/UserDefined跟Action)
username = "jacksugood@gmail.com"
apikey = "gQgpMy3A$uxbRgPuIPL$Rf#^Q&W$oq4"
def myGetFeatureLIST(inputSTR):
    stockResultDICT = articut.parse(inputSTR,userDefinedDictFILE="listed_company.json",  level = 'lv2')
    sentenceLIST = []
    for i in stockResultDICT["result_pos"]:
        if ("ENTITY" in i or "UserDefined" in i) and ("ACTION" in i):
            sentenceLIST.append(i)
    stockNVLIST = [selectNVLIST(i) for i in sentenceLIST]
    return stockNVLIST
    
```

```python
goUpTargetLIST = ["市場","股價","買盤","指數","漲幅","需求","營收","營運","同期","出貨","獲利"]
goUpActionLIST = ["看好","漲","湧入","登上","增","成長", "升","提升","飆"]
def goUpScoreINT(inputSTR):
    stockNVLIST = myGetFeatureLIST(inputSTR)
    goUpINT = 0
    for sent in stockNVLIST:
        for word in sent:
            if any(e in word for e in goUpActionLIST):
                goUpINT+=1
                if any(e in word for e in goUpTargetLIST):
                    goUpINT+=1
    return(goUpINT/len(stockNVLIST))
#設計原理: 我們先依找句子找出如果他有包含跟上漲相關的句子我們就先加上一分，因為有些句子裡頭我們發現只有"漲"加上另一個數字，而這部分沒有被我們
#先前創設的LIST考慮到，所以便先處理一下，而如果一個句子有增加類的動詞以及增加類的名詞的話，我們會更加確定她是跟上漲類的文本相關，因此我們便會
#再多給一分，最後我們因為是以句子為單位下去計分，因此最後用句子數量進行平均。
```

```python
goUpScoreLIST = [goUpScoreINT(i) for i in stockContentLIST]
```

```python
goUpScoreLIST = [round(i,3) for i in goUpScoreLIST]
```

```python
stockDF["go_up_score"] = goUpScoreLIST
stockDF
```
